##데이터 다듬기

```python
# 구 이름과 동 이름은 데이터를 분류하는데 사용되지 않으므로 표에서 사용
train_df.drop(['district'], axis=1, inplace = True) # axis=1 : 열 제거 (axis=0 : 행 제거)
test_df.drop(['dong'], axis=1, inplace = True)      # inplace = True : 데이터 삭제 후 빈 셀 제거

X_train = train_df[['longitude', 'latitude']] # 학습 데이터 중 입력 데이터 뽑기
y_train = train_df[['label']]                 # 학습 데이터 중 레이블(출력,정답 데이터) 뽑기

X_test = test_df[['longitude', 'latitude']] #테스트 데이터 중 입력 데이터 뽑기
y_test = test_df[['label']]                 #테스트 데이터 중 레이블(출력,정답 데이터) 뽑기
```

##모델 학습

```python
from sklearn import tree, preprocessing
import numpy as np

le = preprocessing.LabelEncoder() # LabelEncoder로 레이블을 숫자로 변경하기 위해 설정
y_encoded = le.fit_transform(y_train) # y_train(레이블)의 강북~강서를 0~3의 숫자로 변경

clf = tree.DecisionTreeClassifier(random_state=35).fit(X_train, y_encoded) # 사이킷런의 의사결정트리 학습 실행
#random_state : 난수를 발생시킬 때 사용하는 시드(seed) 값. 이 값이 동일한 경우 같은 순서의 난수가 발생함.
```

##파라미터 없이 학습한 의사결정 트리 시각화 -> 과대적합(Overfitting) 발생

```python
def display_decision_surface(clf,X, y): #사용자 정의 함수(clf:의사결정트리 객체, X: 입력, y: 출력 레이블)
    
    # 챠트의 범위가 모든 학습 데이터를 포함하도록 설정
    x_min = X.longitude.min() - 0.01
    x_max = X.longitude.max() + 0.01
    y_min = X.latitude.min() - 0.01 # 여기서 y의 의미는 레이블이 아님. y축을 의미함. 주의 !!
    y_max = X.latitude.max() + 0.01
    
    # 파라미터 설정
    n_classes = len(le.classes_) # 이 경우 n_classes에는 4가 저장됨(강북~강서)
    plot_colors = "rywb"         # red,yellow,white,blue
    plot_step = 0.001            # 구간 간격

    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),np.arange(y_min, y_max, plot_step))
                        # np.arange(시작, 끝, 값 사이의 간격) : 시작부터 끝 사이의 간격이 일정한 값 반환
    # 학습 데이터의 경우, 이 문장 실행 후, xx = [[126.8395, 126.8405, ...... , 127.1595, 127.1605],  <= 우측으로 증가
    #                                           [126.8395, 126.8405, ...... , 127.1595, 127.1605],
    #                                                         .....................
    #                                           [126.8395, 126.8405, ...... , 127.1595, 127.1605]] 
    #                                      yy = [[37.4419, 37.4419, ...... , 37.4419, 37.4419],  <= 아래로 증가
    #                                           [37.4429, 37.4429, ...... , 37.4429, 37.4429],
    #                                           [37.4439, 37.4439, ...... , 37.4439, 37.4439],
    #                                                         .....................
    #                                           [37.6779, 37.6779, ...... , 37.6779, 37.6779]]   

    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # xx.shape : xx의 개수
    # Z.reshape(xx.shape) : Z를 ()의 개수만큼 2차원 리스트로 만들어줌. 이거를 하는 이유는 clf.predict의 결과가 저장된 Z는 
    #                       1차원 리스트로 저장되어 있음. 이거를 2차원 그래프로 나타내기 위해 reshape 명령을 사용해서 
    #                       2차원 리스트로 만들어 주는 것임.

    # Z 값(예측 결과)에 따른 색상을 그림에 표시
    plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)
                     # x좌표, y좌표, Z는 그려질 데이터(즉, 예측 결과 강서~강북을 나타내는 0~3).              
                     # cmap의 RdYlBu 색상에서 노란색(Yl)과 파란색(Bu) 사이에 흰색(White)가 자동 설정됨.
                     # 색상은 위에 있는 plot_colors와 순서가 동일하게 설정.

    # 학습 데이터를 챠트에 표시
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i) #y(출력 레이블) 리스트에서 i가 있는 위치 찾기
        plt.scatter(X.loc[idx].longitude, X.loc[idx].latitude, c=color, label=le.classes_[i], # i의 레이블 이름 저장. 즉 강서~강북
                    cmap=plt.cm.RdYlBu, edgecolor='black', s=200) # s: 크기
    #zip(a,b) : a와 b를 합쳐서 순서대로 튜플형으로 반환. (0, Red),(1, Yellow),(2, White),(3, Blue)


    plt.title("Decision surface of a decision tree",fontsize=16) # 챠트 제목
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=14) # 챠트 기호 설명(범례(legend) 만들기)
             # bbox_to_anchor : 차트에서의 범례 상대적 위치, (0,0): 좌측 하단, (0,1): 좌측 상단, (1,0): 우측 하단, (1,1): 우측 상단
             # loc : legend(범례) 박스의 기준점. 1: upper right, 2: upper left, 3: lower left, 4: lower right, 5: right, 6: center left
             #                                   7: center right, 8: lower center, 9: upper center, 10: center
             # borderaxespad : legend(범례) 박스와 기준 축 사이의 간격, 0으로 해야 기준축에 딱 붙게 됨.

    
    plt.xlabel('longitude',fontsize=16) # x축의 이름과 폰트 크기 설정
    plt.ylabel('latitude',fontsize=16) # y축의 이름과 폰트 크기 설정

    plt.rcParams["figure.figsize"] = [7,5]     # 챠트 크기 설정
    plt.rcParams["xtick.labelsize"] = 14 # x축 좌표상의 폰트 크기 설정
    plt.rcParams["ytick.labelsize"] = 14 # y축 좌표상의 폰트 크기 설정
    plt.show() # 챠트 그리기

display_decision_surface(clf,X_train, y_encoded) # 함수실행
# 결과: 아무 파라미터 설정 없이 의사결정트리 학습을 하면 다음 그림과 같이 과대적합(Overfitting)이 발생 
```

##과대적합(Overfitting)을 피하기 위해 별도의 파라미터를 설정한 경우

```python
clf = tree.DecisionTreeClassifier(max_depth=4,          #트리의 최대 한도 깊이
                                  min_samples_split=2,  #자식 노드를 갖기 위해(=아래로 분할되기 위해)노드가 가져야 할 최소 데이터 수
                                  min_samples_leaf=2,   #leaf노드(자식 노드가 없는 노드)가 가지고 있어야 할 최소 데이터 수
                                  random_state=70).fit(X_train, y_encoded.ravel()) #random_state : 위에서 설명했음.
display_decision_surface(clf,X_train, y_encoded)                # y_encoded 뒤에 .rabel() 필요없음(교재 틀림)
```

##graphviz를 사용한 의사결정트리 시각화

```python
import graphviz

# 기본 데이터만 시각화(특징 값, 지니 계수, 샘플, 값) -> 그림이 별로..비추천..
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("seoul") # 구글 가상머신에 seoul이라는 텍스트 파일과 pdf 파일 생성
graph # 화면에 Display
```

```python
# 추가 데이터로 시각화(특정 값, 지니 계수, 샘플, 값, 클래스)
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=['longitude','latitude'],  
                         class_names=['Gangbuk', 'Gangdong', 'Gangnam', 'Gangseo'],  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)
graph.render("seoul2") # 구글 가상머신에 seoul2라는 텍스트 파일과 pdf 파일 생성
graph # 화면에 Display
```

##모델 테스트

```python
from sklearn.metrics import accuracy_score

pred = clf.predict(X_test) # 테스트 데이터로 예측 실행. 예측 결과는 pred에 0~3의 숫자로 저장됨.
print("accuracy : " + str( accuracy_score(y_test.values.ravel(), le.classes_[pred]) * 100)+'%' )
                                                                # 숫자로 된 pred의 결과를 le.classes_[pred]를
                                                                # 이용하여 레이블 이름, 즉 강서~강북으로 변경함.
```
